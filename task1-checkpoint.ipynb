```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame called 'data' with a 'gender' column
data = pd.DataFrame({
    'gender': ['Male', 'Female', 'Female', 'Male', 'Female', 'Male', 'Male', 'Female']
})

# Count the occurrences of each gender
gender_counts = data['gender'].value_counts()

# Create a bar chart
gender_counts.plot(kind='bar', color=['blue', 'pink'])
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame called 'data' with an 'age' column
data = pd.DataFrame({
    'age': [23, 45, 31, 35, 25, 42, 28, 39, 22, 33, 40, 30, 27, 50]
})

# Create a histogram
plt.hist(data['age'], bins=5, color='red', edgecolor='grey')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
